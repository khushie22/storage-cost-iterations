```python
import pandas as pd
```

```python
# Load the data and show one row per CombinationID (unique combinations)
df = pd.read_csv('storage_cost_matrix_training_data.csv')

# Get one row for each CombinationID
unique_combinations = df.drop_duplicates(subset=['CombinationID'], keep='first')

# Show only the combination-defining columns (excluding cloud provider specific outputs)
combination_cols = [
    'CombinationID',
    'Class',
    'NumberOfDatabases',
    'DatabaseCapacityGB',
    'TierAllocation_Hot_Percent',
    'TierAllocation_Cold_Percent',
    'TierAllocation_Archive_Percent',
    'TierAllocation_Hot_GB',
    'TierAllocation_Cold_GB',
    'TierAllocation_Archive_GB',
    'Azure_Hot_ReadOperations',
    'Azure_Hot_WriteOperations',
    'Azure_Hot_QueryAccelerationScannedGB',
    'Azure_Hot_QueryAccelerationReturnedGB',
    'Azure_Hot_IterativeReadOperations',
    'Azure_Hot_IterativeWriteOperations',
    'Azure_Hot_OtherOperations',
    'Azure_Cold_ReadOperations',
    'Azure_Cold_WriteOperations',
    'Azure_Cold_QueryAccelerationScannedGB',
    'Azure_Cold_QueryAccelerationReturnedGB',
    'Azure_Cold_IterativeWriteOperations',
    'Azure_Cold_DataRetrievalGB',
    'Azure_Cold_StorageDurationDays',
    'Azure_Archive_ReadOperations',
    'Azure_Archive_WriteOperations',
    'Azure_Archive_ArchiveHighPriorityRead',
    'Azure_Archive_ArchiveHighPriorityRetrievalGB',
    'Azure_Archive_DataRetrievalGB',
    'Azure_Archive_IterativeWriteOperations',
    'Azure_Archive_StorageDurationDays',
    'AWS_Hot_PutCopyPostListRequests',
    'AWS_Hot_GetSelectRequests',
    'AWS_Cold_PutCopyPostListRequests',
    'AWS_Cold_GetSelectRequests',
    'AWS_Cold_DataRetrievalGB',
    'AWS_Cold_StorageDurationDays',
    'AWS_Archive_PutCopyPostListRequests',
    'AWS_Archive_GetSelectRequests',
    'AWS_Archive_DataRetrievalGB',
    'AWS_Archive_DataRetrievalRequests',
    'AWS_Archive_RetrievalType',
    'AWS_Archive_StorageDurationDays'
]

# Display the unique combinations
unique_combinations[combination_cols]
```

```python
# Filter unique combinations by DatabaseCapacityGB
# Set your desired DatabaseCapacityGB value here
capacity_gb = 13.33  # Change this to your desired value

# Create filtered dataframe
filtered_df = unique_combinations[unique_combinations['DatabaseCapacityGB'] == capacity_gb].copy()

# Display the filtered dataframe
x = filtered_df[combination_cols]
```

```python
# Display complete dataframe without truncation
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# Show the complete dataframe x
x
```

```python
# Show all unique tier allocation combinations
tier_allocation_cols = [
    'TierAllocation_Hot_Percent',
    'TierAllocation_Cold_Percent',
    'TierAllocation_Archive_Percent'
]

# Get unique combinations of tier allocation
unique_tier_allocations = unique_combinations[tier_allocation_cols].drop_duplicates().sort_values(by=['TierAllocation_Hot_Percent', 'TierAllocation_Cold_Percent', 'TierAllocation_Archive_Percent'])

# Display all unique tier allocation combinations
unique_tier_allocations
```
